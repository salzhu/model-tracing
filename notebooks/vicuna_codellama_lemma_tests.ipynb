```python
import numpy as np
import matplotlib.pyplot as plt

steps = np.arange(0, 1.1, step=0.1)

codellama_llemma_github = [
    1.2996032441534646,
    1.2460176603096287,
    1.217616707086563,
    1.2116210326188948,
    1.2163302716685505,
    1.2220258407476472,
    1.233115603284138,
    1.2432246815140655,
    1.254131184481993,
    1.274987205499556,
    1.2807087647478754,
]
vicuna_llemma_github = [
    1.5227982271008376,
    1.498126043779094,
    1.5205418958169659,
    1.5046829595798399,
    1.4619252783496206,
    1.3999936340785608,
    1.3469996423256108,
    1.2997660306168766,
    1.2615722025313028,
    1.2466192132816083,
    1.2807087647478754,
]
vicuna_codellama_github = [
    1.5227982271008376,
    1.5039317066349633,
    1.4965518989214084,
    1.4627520845430653,
    1.4363810685349674,
    1.4123091766747033,
    1.3837488606208708,
    1.3500255462599964,
    1.3174012503245982,
    1.3034572735792254,
    1.2996032441534646,
]

# Calculate second derivative approximation for each curve


def second_derivative(x, y):
    dx = x[1] - x[0]
    first_derivative = np.diff(y) / dx
    second_derivative = np.diff(first_derivative) / dx
    return second_derivative


codellama_llemma_second_derivative = second_derivative(steps, codellama_llemma_github)
vicuna_llemma_second_derivative = second_derivative(steps, vicuna_llemma_github)
vicuna_codellama_second_derivative = second_derivative(steps, vicuna_codellama_github)

# Plot the original curves
plt.figure(figsize=(8, 6))
plt.plot(steps, codellama_llemma_github, label="Codellama --> Llemma")
plt.plot(steps, vicuna_llemma_github, label="Vicuna --> LLema")
plt.plot(steps, vicuna_codellama_github, label="Vicuna --> Codellama")

# Plot the second derivative approximations
plt.plot(
    steps[:-2], codellama_llemma_second_derivative, label="Codellama --> Llemma (2nd Derivative)"
)
plt.plot(steps[:-2], vicuna_llemma_second_derivative, label="Vicuna --> LLema (2nd Derivative)")
plt.plot(
    steps[:-2], vicuna_codellama_second_derivative, label="Vicuna --> Codellama (2nd Derivative)"
)

plt.title("Github dataset")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend(fontsize="8")
plt.grid(True)
plt.show()
```

```python
# my calculations

codellama_llemma_github_normalized_step1 = []
for i in range(11):
    codellama_llemma_github_normalized_step1.append(
        codellama_llemma_github[i] - codellama_llemma_github[0]
    )

plt.plot(steps, codellama_llemma_github_normalized_step1, "-")
plt.title("Step 1: Subtract first value")
plt.ylabel("Loss (normalized)")
plt.show()
```

```python
total_change = codellama_llemma_github[-1] - codellama_llemma_github[0]
print(f"Total change: {total_change}")

codellama_llemma_github_normalized_step3 = []
for i in range(11):
    delta = total_change * i / 10
    print(f"Delta: {delta}")
    codellama_llemma_github_normalized_step3.append(
        codellama_llemma_github[i] - codellama_llemma_github[0] - delta
    )

plt.plot(steps, codellama_llemma_github_normalized_step3, "-")
plt.title("Step 3: Subtract linearly interpolated change")
plt.ylabel("Loss (normalized)")
plt.show()
```

```python
codellama_llemma_github_normalized = []
vicuna_llemma_github_normalized = []
vicuna_codellama_github_normalized = []
for i in range(11):
    codellama_llemma_github_normalized.append(
        codellama_llemma_github[i]
        - (codellama_llemma_github[-1] - codellama_llemma_github[0]) * i / 10
        - codellama_llemma_github[0]
    )
    vicuna_llemma_github_normalized.append(
        vicuna_llemma_github[i]
        - (vicuna_llemma_github[-1] - vicuna_llemma_github[0]) * i / 10
        - vicuna_llemma_github[0]
    )
    vicuna_codellama_github_normalized.append(
        vicuna_codellama_github[i]
        - (vicuna_codellama_github[-1] - vicuna_codellama_github[0]) * i / 10
        - vicuna_codellama_github[0]
    )
```

```python
plt.plot(steps, codellama_llemma_github_normalized, "-")
plt.plot(steps, vicuna_llemma_github_normalized, "-")
plt.plot(steps, vicuna_codellama_github_normalized, "-")

plt.title("Normalized endpoints on Github dataset")
plt.ylabel("Loss (normalized)")

plt.legend(
    ["Codellama --> Llemma", "Vicuna --> LLema", "Vicuna --> Codellama"],
    loc="upper right",
    fontsize="6",
)
```
